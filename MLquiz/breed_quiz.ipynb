```python
from sklearn import tree
import matplotlib.pyplot as plt
import pandas as pd
```

```python
breed= pd.read_csv('pet.csv')
```

```python
breed
```

```python

x=breed.drop(['breed', 'description','temperament','group','grooming_frequency_category','shedding_category','energy_level_category','trainability_category','demeanor_category'],axis=1)
y=breed['breed']
```

```python
x
```

```python
y
```

```python
dt = tree.DecisionTreeClassifier()
dt.fit(x, y)
```

```python
plt.figure(figsize = (50,50))
tree.plot_tree(dt, filled = True, fontsize = 12)
```

```python
dt.predict(x)
```

```python
from sklearn.metrics import accuracy_score
xi = dt.predict(x)
yi = y
accuracy = accuracy_score(yi, xi)

# Print the accuracy
print("Accuracy:", accuracy)
```

```python
from joblib import dump, load
dump(dt, 'breedquiz.joblib')
```

```python
clf = load('breedquiz.joblib')
```
